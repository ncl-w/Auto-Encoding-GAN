```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA

# --------- Parameters ----------
n_samples = 3000       # total samples
n_features = 12        # number of dimensions
n_clusters = 5         # number of clusters
cluster_std = 2      # standard deviation of each cluster

# --------- Generate synthetic data ----------
X, y = make_blobs(n_samples=n_samples,
                  n_features=n_features,
                  centers=n_clusters,
                  cluster_std=cluster_std,
                  random_state=42)

print(f"Generated shape: {X.shape}")  # (3000, 12)

# --------- Reduce to 2D using PCA ----------
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X)

# --------- Plot ----------
plt.figure(figsize=(6,6))
scatter = plt.scatter(X_2d[:, 0], X_2d[:, 1], c=y, cmap='tab10', alpha=0.6, s=10)
plt.title(f"{n_clusters}-Cluster GMM (Projected to 2D)")
plt.xlabel("PCA-1")
plt.ylabel("PCA-2")
plt.grid(True)
plt.show()
```

```python
multi_d = np.hstack([X, y.reshape((-1, 1))])
np.savetxt("multiD.txt", multi_d, fmt="%d", delimiter=" ")
```
